```python
# Import sales_data.csv into new_customer from CSV (PySpark-only solution)
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Update path as needed
csv_path = '/Workspace/Users/khajaparveez14@gmail.com/Data-Engineering-Bootcamp/sales_data.csv'

df = (spark.read
    .option('header', 'true')
    .option('inferSchema', 'true')
    .csv(csv_path)
)
# Standardize column names for case-sensitive Delta (fix as needed)
df = df.select(
    col('sale_id').cast('int').alias('Sales_id'),
    col('customer_id').cast('string').alias('customer_id'),
    col('product').cast('string').alias('product'),
    col('amount').cast('double').alias('amount'),
    col('sale_date').cast('date').alias('sale_date'),
    col('region').cast('string').alias('region')
)

df.write.format('delta').mode('append').saveAsTable('new_customer')
print('CSV data appended to new_customer successfully.')
```
